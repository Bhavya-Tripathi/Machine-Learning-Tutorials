```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.stats import norm
random.seed(0)
mu = 5
sigma = 1
s = np.random.normal(mu, sigma, size = (1,100))
x = random.sample(list(s[0]),10)

plt.subplot(211)
plt.title("Gaussian Distribution of 10 samples with \u03BC=5 and \u03C3\u00b2=1")
plt.xlabel("X")
plt.ylabel("p(x)")
plt.scatter(x, norm.pdf(x,5,1))
mu = range(0, 11)
def get_likelihood(x, mu):
    ll = [0]*len(mu)
    for mu_val in  mu:
        for i in range(len(x)):
            ll[mu_val] += norm.pdf(x[i], mu_val,1)
    return ll
ll = get_likelihood(x, mu)
plt.subplot(212)
plt.title("Likelihood function for \u03BC=1 to \u03BC=10")
plt.xlabel("\u03BC")
plt.ylabel("log likelihood")
plt.plot(mu, ll, color='r')
plt.tight_layout()
plt.show()
```
